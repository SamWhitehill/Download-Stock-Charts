```python
import urllib.request
import pandas as pd
from PIL import Image

sTicker ="DOX"
start ="7/27/1998"
end ="7/27/1999"
lStartYr =2009
lEndYr = 2013



def fnSaveChart(pTicker, pStart, pEnd,  pFilePath=""):
    '''
    params:
    pTicker = symbol of stock e.g., MSFT, INTC, IBM
    pStart = starting year of the stock trading history
    pEnd = ending year of the stock trading history
    pMove = boolean indicating whether the last trade closed up from the prior
    day.

    action:
    saves a png file from bigcharts api using params above

    '''

    sTicker =pTicker
    start =str(pStart)
    end =str(pEnd)
    
    sURL="https://api.wsj.net/api/kaavio/charts/big.chart?nosettings=1&symb="+sTicker+"&uf=0&type=4"
    sURL=sURL+"&size=4&style=320&freq=1&entitlementtoken"
    sURL=sURL+"=0c33378313484ba9b46b8e24ded87dd6&"
    sURL=sURL+"startdate="+start+"&enddate="+end
    sURL=sURL+"&rand=221603175&compidx=aaaaa%3a0&ma=4&maval=20,50,200&lf=1&lf2=0&lf3=0&height=50&width=50&mocktick=1"

    lStart=start.replace("/","_")
    lEnd =end.replace("/","_")

    try:

        lFileName = pFilePath+sTicker+"_"+lStart+"_"+lEnd+".png"
        urllib.request.urlretrieve(sURL, lFileName)



        im = Image.open(lFileName)   
        width, height = im.size 
          
        # Setting the points for cropped image 
        left = 1
        top = height +20
        right = 1
        bottom =1 # 3 * height / 4
          
        # Cropped image of above dimension 
        # (It will not change orginal image) 
        im1 = im.crop((0, 16, width, height-12))
        #im1.show()
        print ('Downloaded Chart to:',lFileName)
        im1.save(lFileName)

    except Exception as e:
        print(e)


#FILE_PATH ="C:\\Stock Charts\\NASDAQ Charts\\Train\\"

FILE_PATH ="C:\\Stock Charts\\Charts - No Price\\New Validation 2006 to 2009\\"


sTicker ="MSFT"
start ="8/27/2023"
end ="8/27/2024"
lStartYr =2009
lEndYr = 2013

#Example call to download stock chart image to the hard drive
FILE_PATH ="c:\\temp\\"
fnSaveChart(sTicker, lStartYr, lEndYr, pFilePath=FILE_PATH)
```
